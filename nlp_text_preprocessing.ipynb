```python
import nltk 
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer , WordNetLemmatizer 
from nltk.corpus import stopwords 
import re 
import pandas as pd 
```

```python
data= pd.read_csv('tripadvisor_hotel_reviews.csv') 
data.head()
```

```python
data['Review'][0]
```

```python
data['lower_review']=data['Review'].str.lower() 
data.head()
```

```python
stop_word= stopwords.words('english') 
data['no_stopword_review']= data['lower_review'].apply(lambda x : " ".join([word for word in x.split() if word not in stop_word]))
data.head()
```

```python
data['no_*_review']= data.apply(lambda x : re.sub(r"[*]" , "star" , x['no_stopword_review']), axis=1 )
data.head()
```

```python
data['no_punc_review']= data.apply(lambda x : re.sub(r"[^\w\s]" , "" , x['no_*_review']), axis=1 )
data.head()
```

```python
data['tokenized']=data.apply(lambda x : word_tokenize(x['no_punc_review']) , axis=1 )
data.head()
```

```python
ps=PorterStemmer()
data['stemmed'] = data['tokenized'].apply (lambda tokens  : [ps.stem(token) for token in tokens])
data.head()
```

```python
lemmatizer=WordNetLemmatizer() 
data['lemmatized']= data['stemmed'].apply (lambda  tokens : [lemmatizer.lemmatize(token) for token in tokens] )
data.head()
```

```python
cleaning = sum(data['lemmatized'],[]) 
unigrams_1=(pd.Series(nltk.ngrams(cleaning , 1)).value_counts())
print(unigrams_1)
```

```python
bigrams_1=(pd.Series(nltk.ngrams(cleaning , 2)).value_counts())
print(bigrams_1)
```

```python
trigrams_1=(pd.Series(nltk.ngrams(cleaning , 3)).value_counts())
print(trigrams_1)
```
